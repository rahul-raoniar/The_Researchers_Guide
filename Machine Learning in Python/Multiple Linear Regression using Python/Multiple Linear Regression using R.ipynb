### Fitting Linear Regression

```r
model1 = lm(GPA ~., data = data)
```

```r
summary(model1)
```

#### Plotting separate regression line according to Attendance

```r
ggplot(data = data, aes(x = SAT, y = GPA, color = Attendance))+
geom_point()+
geom_smooth(method = "lm")
```
